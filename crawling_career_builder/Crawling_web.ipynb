```python
import requests
from bs4 import BeautifulSoup
```

```python
# Read html text
url = 'https://careerbuilder.vn/viec-lam/duoc-pham-c7-vi.html'
html_text = requests.get(url).text
html_text
```

```python
tree = BeautifulSoup(html_text, 'html.parser')
```

```python
# Find all job containers
job_containers = tree.find_all("div", {"class": "figure"})
len(job_containers)
```

Kế đến, với mỗi phần tử (mỗi tuyển dụng) trong `job_containers`, ta sẽ rút trích ra các thông tin:
- Công ty tuyển dụng 
- Địa điểm 
- Ngày đăng thông báo
- Lương.
- Tựa đề.
- Link đến nội dung chi tiết.
- Nội dung chi tiết 

Sau đó, lưu các thông tin này xuống file `csv` (mỗi tuyển dụng ứng với một dòng). Để tránh nhập nhằng giữa dấu `,` trong nội dung các thông tin và dấu `,` phân tách các giá trị trong file `csv`, ta sẽ dùng dấu `\t` (tab) để phân tách các giá trị trong file `csv` (file `csv` cho phép làm như vậy, chứ không bắt buộc là phải dùng dấu `,`).

```python
import pandas as pd

df_job = pd.DataFrame(columns = ["company","location","date","salary","title","url"])
df_job.head()
```

```python
dic_job = {}
for job_container in job_containers:
    company = job_container.find('div', {'class': 'caption'}).find('a',{'class': 'company-name'}).string
    location = job_container.find('div', {'class': 'location'}).find('ul').find("li").get_text()
    date = job_container.find('div', {'class': 'time'}).find('time').get_text()
    salary = job_container.find('div', {'class': 'salary'}).get_text()
    title = job_container.find('div', {'class': 'title'}).find("a").get_text().strip()
    url = job_container.find('div', {'class': 'title'}).find("a",{'class':{'job_link'}})['href']
    # add to dictionary for append pandas
    dic_job['company'] = company
    dic_job['location'] = location
    dic_job['date'] = date
    dic_job['salary'] = salary
    dic_job['title'] = title
    dic_job['url'] = url
    df_job=  df_job.append(dic_job,ignore_index=True)
```

```python
df_job.to_csv("JOB.csv", sep = "\t",index=False)
```

```python
def is_number(s):
    
    try:
        float(s)
        return True
    except ValueError:
        return False
    
def min_salary(str_salary, _min = True):
    words= str_salary.split()
    salary = []
    for word in words:
        word = word.replace(",", ".")
        if is_number(word):
            salary.append(float(word))
    if salary and _min:
        return min(salary)
    elif salary and not _min:
        return max(salary)
    else:
        return -1
```

```python
df_job2["min_salary"] = df_job2.apply(lambda row :min_salary(row["salary"]),axis =1)
df_job2["max_salary"] = df_job2.apply(lambda row :min_salary(row["salary"], _min = False),axis =1)
```
